```python
def leer_fichero(ruta):
  with open(ruta, "r", encoding="UTF8") as fich:
    lineas = fich.readlines()
  return lineas
```

```python
def leer_fichero_plano(ruta):
  with open(ruta, "r", encoding="UTF8") as fich:
    lineas = fich.read()
  return lineas
```

```python
def escribir_fichero(ruta, texto):
  with open(ruta, "w") as fich:
    fich.write(texto)
```

```python
def quitar_guiones(lista_lineas): #QUITA LOS SALTOS DE LÍNEA Y LOS GUIONES FINALES EN PALABRAS QUE LO LLEVEN
  resultado = ''
  for linea in lista_lineas:
    sin_blanco = linea.strip()
    if sin_blanco.endswith('-'):
      longitud = len(sin_blanco)
      sin_guion = sin_blanco[:longitud -1]
      resultado = resultado + sin_guion
    else:
      resultado = resultado + sin_blanco + ' '
  resultado = resultado.strip()


  return resultado
```

```python
def quitar_folios (texto):
  resultado = ""
  posicion = texto.find("[fol.")
  while posicion != -1:
    principio = texto[:posicion]
    resto = texto[posicion:]
    posicion_cerrado = resto.find("]")
    resto_cerrado = resto[posicion_cerrado +1:]
    resultado = resultado + principio
    texto = resto_cerrado
    posicion = texto.find("[fol.")
  resultado = resultado + texto
  return resultado
```

```python
def quitar_reclamos (texto):
  resultado = ""
  posicion = texto.find("{CW.")
  while posicion != -1:
    principio = texto[:posicion]
    resto = texto[posicion:]
    posicion_cerrado = resto.find("}")
    resto_cerrado = resto[posicion_cerrado +1:]
    resultado = resultado + principio
    texto = resto_cerrado
    posicion = texto.find("{CW.")
  resultado = resultado + texto
  return resultado
```

```python
texto = "{MIN. Ce<lestina> LV<CRECIA> ME<LIBEA> CA<LISTO>}Comedia de: calisto & melibe<<a>>."
def quitar_miniaturas (texto):
  resultado = ""
  posicion = texto.find("{MIN.")
  while posicion != -1:
    principio = texto[:posicion]
    resto = texto[posicion:]
    posicion_cerrado = resto.find("}")
    resto_cerrado = resto[posicion_cerrado +1:]
    resultado = resultado + principio
    texto = resto_cerrado
    posicion = texto.find("{MIN.")
  resultado = resultado + texto
  return resultado
```

```python

def quitar_iniciales (texto):
  resultado = ""
  posicion = texto.find("{IN")
  while posicion != -1:
    principio = texto[:posicion]
    resto = texto[posicion:]
    posicion_cerrado = resto.find("}")
    resto_cerrado = resto[posicion_cerrado + 1:]
    resultado = resultado + principio
    texto = resto_cerrado
    posicion = texto.find("{IN")
  resultado = resultado + texto
  return resultado
```

```python
# quitar_corchetes elimina las inserciones del editor.
def quitar_corchetes (texto):
  resultado = ""
  posicion = texto.find("[")
  while posicion != -1: # mientras encuentre un folio o marca de folio
    principio = texto[:posicion]
    resto = texto[posicion:]
    posicion_cerrado = resto.find("]") #hay que darle instrucciones para que borre el . y el número anterior: 5 y 8
    resto_cerrado = resto[posicion_cerrado + 1:]
    resultado = resultado + principio
    texto = resto_cerrado
    posicion = texto.find("[")
  resultado = resultado + texto
  return resultado
```

```python
import re
def quitar_blancos(texto):
  resultado = " ".join(texto.split())
  return resultado
```

```python
def limpieza_por_reemplazamiento(texto):
  resultado = texto.replace('-', '')
  resultado = resultado.replace('<', '')
  resultado = resultado.replace('>', '')
  resultado = resultado.replace('(', '')
  resultado = resultado.replace('((', '')
  resultado = resultado.replace(')', '')
  resultado = resultado.replace('))', '')
  resultado = resultado.replace('c\'', 'ç')
  resultado = resultado.replace('n~', 'ñ')
  resultado = resultado.replace('z\'', 'z')
  resultado = resultado.replace('{ILL.', '')
  resultado = resultado.replace('{CB1.', '')
  resultado = resultado.replace('% {RUB.', '')
  resultado = resultado.replace('{RUB.', '')
  resultado = resultado.replace('%', '')
  resultado = resultado.replace('{LAT.', '')
  resultado = resultado.replace('}', '')
  return resultado
```

```python
def limpiar(ruta):
  texto = leer_fichero(ruta)
  texto = quitar_guiones(texto)
  texto = quitar_folios(texto)
  texto = quitar_reclamos(texto)
  texto = quitar_miniaturas(texto)
  texto = quitar_iniciales(texto)
  texto = quitar_corchetes(texto)
  texto = limpieza_por_reemplazamiento(texto)
  texto = quitar_blancos(texto)
  texto = texto.strip()
  return texto
```

```python
def letras_equivalentes(pal1, pal2):
  pal1 = pal1.replace("u", "v")
  pal2 = pal2.replace("u", "v")
  pal1 = pal1.replace("i", "j")
  pal2 = pal2.replace("i", "j")
  pal1 = pal1.replace("y", "i")
  pal2 = pal2.replace("y", "i")
  pal1 = pal1.replace("f", "h")
  pal2 = pal2.replace("f", "h")
  pal1 = pal1.replace("b", "v")
  pal2 = pal2.replace("b", "v")
  return pal1 == pal2
```

```python
def sonIguales(pal1, pal2):
  iguales = pal1 == pal2
  return iguales
```

```python
def mayusculas(pal1, pal2):
  pal1_may = pal1.upper()
  pal2_may = pal2.upper()
  iguales = pal1_may == pal2_may
  return iguales
```

```python
def quitar_puntuacion(texto):
  texto = texto.replace(".","")
  texto = texto.replace(",","")
  texto = texto.replace(":","")
  texto = texto.replace(";","")
  texto = texto.replace("/","")
  texto = texto.replace("¿","")
  texto = texto.replace("?","")
  texto = texto.replace("¡","")
  texto = texto.replace("!","")
  texto = texto.replace("'","")
  return texto
```

```python
def buscar_coincidencia (doc1, doc2, i, j):
  i_temp = i
  len1 = len(doc1)
  len2 = len(doc2)
  distancia = 100
  solucion = (False, i, j)
  while i_temp < len1:
    pal1 = doc1[i_temp].text
    j_temp = j
    encontrado = False
    distancia_actual = (i_temp - i) + (j_temp - j)
    while j_temp < len2 and not encontrado and distancia_actual < distancia:
      pal2 = doc2[j_temp].text
      if pal1 == pal2:
        encontrado = True
        distancia = distancia_actual
        solucion = (True, i_temp, j_temp)
      j_temp = j_temp + 1
      distancia_actual = distancia_actual + 1
    i_temp = i_temp + 1
  return solucion
```

```python
def comparar(doc1, doc2):
  i = 0
  j = 0
  len1 = len(doc1)
  len2 = len(doc2)
  len_min = min(len1,len2)
  separado = []
  lista_mayusculas = []
  diferencias_letras_equivalentes = []
  resto_diferencias = []
  diferencias_puntuacion = []
  while i < len_min and j < len_min:
    pal1 = doc1[i].text
    pal2 = doc2[j].text
    if i + 1 < len_min and (pal1 + doc1[i+1].text) == pal2:
      separado.append((pal1,doc1[i+1].text, pal2, i + 1, j + 1))
      i = i + 1
    elif j + 1 < len_min and (pal2 + doc2[j+1].text) == pal1:
      separado.append((pal2,doc2[j+1].text, pal1, i + 1, j + 1))
      j = j + 1
    elif not sonIguales(pal1, pal2) and pal1 in ".,:;/":
     diferencias_puntuacion.append((pal1,pal2,i + 1, j + 1))
     i = i + 1
    elif not sonIguales(pal1, pal2) and pal2 in ".,:;/":
     diferencias_puntuacion.append((pal1,pal2,i + 1, j + 1))
     j = j + 1
    elif not sonIguales(pal1, pal2) and mayusculas(pal1, pal2):
      lista_mayusculas.append((pal1, pal2, i, j))
    elif not sonIguales(pal1, pal2) and letras_equivalentes(pal1, pal2):
      diferencias_letras_equivalentes.append((pal1,pal2, i + 1, j + 1))
    elif not sonIguales(pal1, pal2) and not letras_equivalentes(pal1, pal2):
      resto_diferencias.append((pal1,pal2, i + 1, j + 1))
    i = i + 1
    j = j + 1
  return [separado, lista_mayusculas, diferencias_letras_equivalentes, diferencias_puntuacion, resto_diferencias]
```

```python
def texto_puntuacion_origen(doc, ldoc, i, lj, wit, lwit, punt, sem):
  pal_orig = doc[i].text
  texto = {pal_orig : [wit]}
  for k in range(len(ldoc)):
    pal = ldoc[k][lj[k]].text
    if pal in punt:
      if pal in texto:
        texto[pal].append(lwit[k])
      else:
        texto[pal] = [lwit[k]]
    else:
      if "" in texto:
        texto[""].append(lwit[k])
      else:
        texto[""] = [lwit[k]]
  lista_orig = texto[pal_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + pal_orig + "</lem>"
  del texto[pal_orig]
  rdg = ""
  for clave, valor in texto.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="puntuacion" subtype="presencia_en_base">' + lem + rdg + "</app>"
  return app
```

```python
def sumar_puntuacion(ldoc, lj, puntuacion):
  for k in range(len(ldoc)):
    if ldoc[k][lj[k]].text in puntuacion:
      lj[k] = lj[k] + 1
```

```python
def puntuacion_otros(ldoc, lj, puntuacion):
  hay_puntuacion = False
  for k in range(len(ldoc)):
    hay_puntuacion = hay_puntuacion or ldoc[k][lj[k]].text in puntuacion
  return hay_puntuacion
```

```python
# Asumimos que no hay puntuación en lema
def texto_puntuacion_otros(ldoc, lj, wit, lwit, punt, sem):
  texto = {"" : [wit]}
  for k in range(len(ldoc)):
    pal = ldoc[k][lj[k]].text
    if pal in punt:
      if pal in texto:
        texto[pal].append(lwit[k])
      else:
        texto[pal] = [lwit[k]]
    else:
      texto[""].append(lwit[k])
  lista_orig = texto[""]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '"></lem>'
  del texto[""]
  rdg = ""
  for clave, valor in texto.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="puntuacion" subtype="omision_en_base">' + lem + rdg + "</app>"
  return app
```

```python
def mayusculas_lista(doc1, ldoc, i, lj):
  pal_orig = doc1[i].text.upper()
  k = 0
  iguales = True
  while iguales and k < len(ldoc):
    pal = ldoc[k][lj[k]].text.upper()
    iguales = pal_orig == pal
    k = k + 1
  return iguales
```

```python
def texto_mayusculas(doc, ldoc, i, lj, wit, lwit, sem):
  pal_orig = doc[i].text
  palabras = {pal_orig : [wit]}
  for k in range(len(ldoc)):
    pal = ldoc[k][lj[k]].text
    if pal in palabras:
      palabras[pal].append(lwit[k])
    else:
      palabras[pal] = [lwit[k]]
  lista_orig = palabras[pal_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + pal_orig + "</lem>"
  del palabras[pal_orig]
  rdg = ""
  for clave, valor in palabras.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="grafia" subtype="mayusculas">' + lem + rdg + "</app>"
  return app
```

```python
def letras_equivalentes_lista(doc, ldoc, i, lj):
  pal_orig = reemplazar_equivalentes(doc[i].text)
  equivalentes = True
  k = 0
  while equivalentes and k < len(ldoc):
    actual = reemplazar_equivalentes(ldoc[k][lj[k]].text)
    equivalentes = pal_orig == actual
    k = k + 1
  return equivalentes
```

```python
def reemplazar_equivalentes(pal):
  nueva = pal.replace("u", "v")
  nueva = nueva.replace("i", "j")
  nueva = nueva.replace("y", "i")
  nueva = nueva.replace("f", "h")
  nueva = nueva.replace("b", "v")
  # pal = pal.replace("s", "x")
  return nueva
```

```python
def texto_palabras_equivalentes(doc, ldoc, i, lj, wit, lwit, sem):
  pal_orig = doc[i].text
  palabras = {pal_orig : [wit]}
  for k in range(len(ldoc)):
    pal = ldoc[k][lj[k]].text
    if pal in palabras:
      palabras[pal].append(lwit[k])
    else:
      palabras[pal] = [lwit[k]]
  lista_orig = palabras[pal_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + pal_orig + "</lem>"
  del palabras[pal_orig]
  rdg = ""
  for clave, valor in palabras.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="grafia" subtype="letras_equivalentes">' + lem + rdg + "</app>"
  return app
```

```python
def texto_sin_coincidencia(doc, ldoc, i, lj, wit, lwit, sem):
  pal_orig = doc[i].text
  palabras = {pal_orig : [wit]}
  for k in range(len(ldoc)):
    pal = ldoc[k][lj[k]].text
    if pal in palabras:
      palabras[pal].append(lwit[k])
    else:
      palabras[pal] = [lwit[k]]
  lista_orig = palabras[pal_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + pal_orig + "</lem>"
  del palabras[pal_orig]
  rdg = ""
  for clave, valor in palabras.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="sin_coincidencia">' + lem + rdg + "</app>"
  return app
```

```python
def buscar_coincidencia_varios(doc1, ldoc, i, lj):
  i_temp = i
  len1 = len(doc1)
  len2 = []
  for num in range(len(ldoc)):
    new_len = len(ldoc[num])
    len2.append(new_len)
  distancia = 200
  solucion = (False, i, [])
  while i_temp < len1:
    pal1 = doc1[i_temp].text
    anterior_j = 0
    todos_encontrados = True
    js = [0] * len(ldoc)
    for num in range(len(ldoc)):
      j_temp = lj[num]
      encontrado = False
      distancia_actual = (i_temp - i) + anterior_j + (j_temp - lj[num])
      while j_temp < len2[num] and not encontrado and todos_encontrados and distancia_actual < distancia:
        pal2 = ldoc[num][j_temp].text
        if pal1 == pal2:
          encontrado = True
          js[num] = j_temp
        j_temp = j_temp + 1
        distancia_actual = distancia_actual + 1
      todos_encontrados = todos_encontrados and encontrado
      anterior_j = anterior_j + (j_temp - lj[num] - 1)
    if todos_encontrados and (i_temp - i) + anterior_j < distancia:
      distancia = (i_temp - i) + anterior_j
      solucion = (True, i_temp, js)
    i_temp = i_temp + 1
  return solucion
```

```python
def generar_texto(doc, init, fin):
  i = init
  frase = ""
  while i < fin:
    pal = doc[i].text
    frase = frase + " " + pal
    i = i + 1
  frase = frase.strip()
  return frase
```

```python
def texto_varios(doc, ldoc, i, lj, i_fin, lj_fin, wit, lwit, sem):
  frase_orig = generar_texto(doc, i, i_fin)
  frases = {frase_orig : [wit]}
  for k in range(len(ldoc)):
    frase = generar_texto(ldoc[k], lj[k], lj_fin[k])
    if frase in frases:
      frases[frase].append(lwit[k])
    else:
      frases[frase] = [lwit[k]]
  lista_orig = frases[frase_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + frase_orig + "</lem>"
  del frases[frase_orig]
  rdg = ""
  for clave, valor in frases.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="alternativa">' + lem + rdg + "</app>"
  return app
```

```python
def similaridad_por_testimonio(lista, dicc):
  i = 0
  tam = len(lista)
  while i < tam:
    et1 = lista[i]
    j = i + 1
    while j < tam:
      et2 = lista[j]
      par = (et1, et2)
      if par in dicc:
        dicc[par] = dicc[par] + 1
      else:
        dicc[par] = 1
      j = j + 1
    i = i + 1
```

```python
def insertar_final(doc, ldoc, i, lj, len1, llen, wit, lwit, sem):
  frase_orig = generar_texto(doc, i, len1)
  frases = {frase_orig : [wit]}
  for k in range(len(ldoc)):
    frase = generar_texto(ldoc[k], lj[k], llen[k])
    if frase in frases:
      frases[frase].append(lwit[k])
    else:
      frases[frase] = [lwit[k]]
  lista_orig = frases[frase_orig]
  similaridad_por_testimonio(lista_orig, sem)
  testimonios = " ".join(lista_orig)
  lem = '<lem wit="' + testimonios + '">' + frase_orig + "</lem>"
  del frases[frase_orig]
  rdg = ""
  for clave, valor in frases.items():
    similaridad_por_testimonio(valor, sem)
    testimonios = " ".join(valor)
    rdg = rdg + '<rdg wit="' + testimonios + '">' + clave + "</rdg>"
  app = '<app type="alternativa">' + lem + rdg + "</app>"
  return app
```

```python
def generar_app_lista(doc1, ldoc, wit, lwit, sem):
  i = 0
  lj = [0] * len(ldoc)
  len1 = len(doc1)
  llen = []
  for doc in ldoc:
    tam = len(doc)
    llen.append(tam)
  len_min = min(len1,min(llen))
  texto = ""
  puntuacion = ".,:;/¿?¡!"
  while todos_menores(i, lj, len1, llen):
    pal1 = doc1[i].text
    if todas_iguales(doc1, ldoc, i, lj):
      texto = texto + " " + pal1
      i = i + 1
      sumar1(lj)
    else:
      if separadas_en_origen(doc1, ldoc, i, lj):
        texto_generado = texto_separadas_en_origen(doc1, ldoc, i, lj, wit, lwit, sem)
        texto = texto + " " + texto_generado
        sumar_separadas_en_origen(doc1, ldoc, i, lj)
        i = i + 2
      elif juntas_en_origen(doc1, ldoc, i, lj):
        texto_generado = texto_juntas_en_origen(doc1, ldoc, i, lj, wit, lwit, sem)
        texto = texto + " " + texto_generado
        sumar_juntas_en_origen(doc1, ldoc, i, lj)
        i = i + 1
      elif puntuacion_origen(doc1, i, puntuacion):
        texto_generado = texto_puntuacion_origen(doc1, ldoc, i, lj, wit, lwit, puntuacion, sem)
        texto = texto + " " + texto_generado
        sumar_puntuacion(ldoc, lj, puntuacion)
        i = i + 1
      elif puntuacion_otros(ldoc, lj, puntuacion):
        texto_generado = texto_puntuacion_otros(ldoc, lj, wit, lwit, puntuacion, sem)
        texto = texto + " " + texto_generado
        sumar_puntuacion(ldoc, lj, puntuacion)
      elif mayusculas_lista(doc1, ldoc, i, lj):
        texto_generado = texto_mayusculas(doc1, ldoc, i, lj, wit, lwit, sem)
        texto = texto + " " + texto_generado
        i = i + 1
        sumar1(lj)
      elif letras_equivalentes_lista(doc1, ldoc, i, lj):
        texto_generado = texto_palabras_equivalentes(doc1, ldoc, i, lj, wit, lwit, sem)
        texto = texto + " " + texto_generado
        i = i + 1
        sumar1(lj)
      else:
        (sol, i_fin, lj_fin) = buscar_coincidencia_varios(doc1, ldoc, i, lj)
        if sol:
          texto_generado = texto_varios(doc1, ldoc, i, lj, i_fin, lj_fin, wit, lwit, sem)
          texto = texto + " " + texto_generado
          i = i_fin
          lj = lj_fin
        else:
          texto_generado = texto_sin_coincidencia(doc1, ldoc, i, lj, wit, lwit, sem)
          texto = texto + " " + texto_generado
          i = i + 1
          sumar1(lj)
  if alguno_menor(i, lj, len1, llen):
    texto_generado = insertar_final(doc1, ldoc, i, lj, len1, llen, wit, lwit, sem)
    texto = texto + " " + texto_generado
  texto = texto.strip()
  return texto
```

```python
def quitar_guiones(lista_lineas):
  resultado = ''
  for linea in lista_lineas:
    sin_blanco = linea.strip()
    if sin_blanco.endswith('-'):
      longitud = len(sin_blanco)
      sin_guion = sin_blanco[:longitud -1]
      resultado = resultado + sin_guion
    else:
      resultado = resultado + sin_blanco + ' '
  resultado = resultado.strip()
  resultado = resultado.replace('-', '')
  resultado = resultado.replace('<', '')
  resultado = resultado.replace('>', '')
  resultado = resultado.replace('[*', '')  #[*] Editorial reconstruction of illegible or missing text
  resultado = resultado.replace(']', '') # resto de la reconstrucción, ¿quitaría lo de los folios?
  resultado = resultado.replace('[', '') # al quitar corchetes se mostrarían las adiciones editorial
  #resultado = resultado.replace('s\'', 's')en nuevo txt mp no aparece
  #resultado = resultado.replace('z\'', 'z')en nuevo txt mp no aparece
  #resultado = resultado.replace('s', 'ss') #sustituir "s" por "ss" cuando sea intervocálica, ¿cómo se haría?
  #resultado = resultado.replace(' ', '') #sustituir dos espacios por uno, borra todos los espacios
 # resultado = resultado.replace('&', 'y') #por si hay que sustituir "&" por "y"
  return resultado
```

```python
def sustituir(frase, patron, sust):
  busqueda = patron.search(frase)
  while busqueda != None:
    pat = busqueda[0]
    pat = pat.replace("s'", sust)
    frase = patron.sub(pat, frase,count=1)
    busqueda = patron.search(frase)
  return frase

  #sustituir s' en distintas fases en función de lo que acompañe
```
